```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df_data = pd.read_csv('/content/data (1).csv')
df_data
```

```python
Y = df_data["target"]
df_data = df_data.drop('target', axis=1)
df_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data, Y, test_size=0.2, random_state=42)
x_train1,x_val,y_train1,y_val=train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

#ann on multclass

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
j=15
point=[]
accc=[]
while j<50:
  point.append(j)
  classifier2 = Sequential()
  classifier2.add(Dense(j, activation = 'relu', input_dim =50))
  classifier2.add(Dense(j, activation = 'relu'))
  classifier2.add(Dense(4,activation = 'softmax'))
  classifier2.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
  classifier2.fit(x_train1, y_train1, batch_size = 100, epochs = 100)
  y_pred_2 = classifier2.predict(x_val)
  y_pred2_2=[]
  for i in range(len(y_pred_2)):
    m=max(y_pred_2[i])
    a=np.where(y_pred_2[i]==m)[0][0]
    y_pred2_2.append(a)
  y_pred2_2=np.array(y_pred2_2)
  from sklearn.metrics import accuracy_score
  acc=accuracy_score(y_val,y_pred2_2)
  accc.append(acc)
  j+=5
```

```python
plt.plot(point,accc)
print(max(accc))
l=accc.index(max(accc))
print(point[l])
```
